```python
import random
import csv

facture = ["electricité","eau","téléphone"]

dataset = []

for _ in range(5000):
    montant = random.randint(10, 1000)
    facture = random.choice(["electricité","eau","téléphone"])
    phrase_principale = random.choice(["Paye une facture de", "Effectue le paiement d'une facture de", "Règle la facture de"])
    phrase = f"{phrase_principale} <NOM>{facture}</NOM> de {montant} euros."
    data = {
        "phrase": phrase,
        "facture": facture,
        "montant": montant
    }
    dataset.append(data)

with open('dataset_factures.csv', 'w', newline='') as f:
    fieldnames = ['phrase', 'facture', 'montant']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for data in dataset:
        writer.writerow(data)
```

```python
import pandas as pd

data = pd.read_csv('C:/Users/HP/dataset_factures.csv', encoding='ISO-8859-1')

X = data['phrase']
y_facture = data['facture']
y_montant = data['montant']
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_facture_train, y_facture_test, y_montant_train, y_montant_test = train_test_split(
    X, y_facture, y_montant, test_size=0.2, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC

pipeline_facture = Pipeline([
    ('vect', CountVectorizer()),
    ('clf', LinearSVC())
])

pipeline_montant = Pipeline([
    ('vect', CountVectorizer()),
    ('clf', LinearSVC())
])

pipeline_facture.fit(X_train, y_facture_train)
pipeline_montant.fit(X_train, y_montant_train)
```

```python
accuracy_facture = pipeline_facture.score(X_test, y_facture_test)
accuracy_montant = pipeline_montant.score(X_test, y_montant_test)

print(f"Précision du modèle pour le type de facture : {accuracy_facture}")
print(f"Précision du modèle pour le montant : {accuracy_montant}")
```

```python
phrase1 = "Paye une facture d'electricité de 500 euros."
facture1 = pipeline_facture.predict([phrase1])[0]
montant1 = pipeline_montant.predict([phrase1])[0]

print(f"la facture de la phrase 1 : {facture1}")
print(f"Montant du  de la facture 1 : {montant1} euros")

phrase2 = "Effectue le paiement d'une facture de téléphone de 200 euros."
facture2 = pipeline_facture.predict([phrase2])[0]
montant2 = pipeline_montant.predict([phrase2])[0]

print(f"la facture de la phrase 2 : {facture2}")
print(f"Montant de la facture 2 : {montant2} euros")
```
